### 13.Write the code to implement the following descriptive statistics operations any of the created data frame a. Mean, Median and mode b. Standard Deviations c. Variance d. Kurtosis e. Skewness

```python
import pandas as pd
# Create DataFrame
data = {'values': [10, 12, 15, 18, 20, 22, 25, 30, 35, 40]}
df = pd.DataFrame(data)
print("Dataset:")
print(df)
# a) Mean, Median, Mode
print("\nMean:", df['values'].mean())
print("Median:", df['values'].median())
print("Mode:\n", df['values'].mode())
# b) Standard Deviation
print("\nStandard Deviation:", df['values'].std())
# c) Variance
print("Variance:", df['values'].var())
# d) Kurtosis
print("Kurtosis:", df['values'].kurt())
# e) Skewness
print("Skewness:", df['values'].skew())
```
