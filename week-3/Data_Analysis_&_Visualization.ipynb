# **Data Analysis & Visualization using Pandas, NumPy, Matplotlib**

```python
# Data Analysis & Visualization using Pandas, NumPy, Matplotlib
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load Iris dataset (public CSV link)
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
df = pd.read_csv(url)

print("Dataset (first 10 rows):")
print(df.head(10))

# Select the petal length column
data = df["petal_length"]

print("\nSummary Statistics:")
print(data.describe())

print("\nVariance:", np.var(data, ddof=1))
print("Standard Deviation:", np.std(data, ddof=1))

# Plot histogram
plt.figure(figsize=(8,5))
plt.hist(data, bins=10, color='lightgreen', edgecolor='black')

# Mean line
plt.axvline(np.mean(data), color='red', linestyle='dashed',
            linewidth=2, label=f"Mean: {np.mean(data):.2f}")

# Median line
plt.axvline(np.median(data), color='blue', linestyle='dashed',
            linewidth=2, label=f"Median: {np.median(data):.2f}")

plt.xlabel("Petal Length (cm)")
plt.ylabel("Frequency")
plt.title("Distribution of Petal Length")
plt.legend()
plt.show()

# Skewness inference
mean_value = np.mean(data)
median_value = np.median(data)

print("\nInference:")
print("Mean value:", mean_value)
print("Median value:", median_value)

if mean_value > median_value:
    print("Data is slightly right-skewed")
elif mean_value < median_value:
    print("Data is slightly left-skewed")
else:
    print("Data is symmetric")
```
