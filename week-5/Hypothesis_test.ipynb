##22.write a code to test for hypothesis,if test for Z is accepted or rejected.

```python
#Hypothesis Test
import math
n=100
x=60
p_hat=x/n
SE=math.sqrt(p_hat*(1-p_hat)/n)
#Confidence interval
z=1.96
lower=p_hat-z*SE
upper=p_hat+z*SE
print("Sample proportion:",p_hat)
print("Confidence Interval=",p_hat)
po=0.50
z_test=(p_hat-po)/math.sqrt(po*(1-po)/n)
print("Z-value:",z_test)
if abs(z_test)>1.96:
  print("Reject Ho")
else:
  print("Do not reject Ho")
```
