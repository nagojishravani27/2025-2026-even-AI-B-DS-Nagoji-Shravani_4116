## 24.Write a code to implement Linear regression with an example URL

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
#Dateset URL
url="https://raw.githubusercontent.com/ybiFoundation/Dataset/main/Salary%20Data.csv"
#Load dataset
data=pd.read_csv(url)
#Display first rows
print("Dataset Preview")
print(data.head())
#Define features and target
X=data[['Experience Years']]
y=data['Salary']
#create model
model=LinearRegression()
# Train model
model.fit(X,y)
#Model Parameters
print("\n Model parameters")
print("Slope:",model.coef_[0])
print("Intercept:",model.intercept_)
#Prediction example
years=[[5]]
predicted_salary=model.predict(years)
print("Predicted Salary for 5 years Experience:",predicted_salary[0])
#predictions for dataset
y_pred=model.predict(X)
#calculate error
mse=mean_squared_error(y,y_pred)
print("\n Mean Squared Error:",mse)
#Visualization
plt.scatter(X,y)
plt.plot(X,y_pred)
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("LinearRegression Salary Prediction")
plt.show()
```
