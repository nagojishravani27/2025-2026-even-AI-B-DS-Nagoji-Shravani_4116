**Mean,Median,Variance,Standard Deviations, Kurtosis,Skewness**

```python
import pandas as pd
import numpy as np

# Create DataFrame
data = [10, 12, 15, 18, 20, 22, 25, 30, 35, 40]
df = pd.DataFrame(data, columns=["values"])

print("Dataset:")
print(df)

# Descriptive Statistics
mean_value = df["values"].mean()
median_value = df["values"].median()
mode_value = df["values"].mode()
variance_value = df["values"].var()
std_dev_value = df["values"].std()
skewness_value = df["values"].skew()
kurtosis_value = df["values"].kurt()

print("\nDescriptive Statistics:")
print("Mean:", mean_value)
print("Median:", median_value)
print("Mode:", list(mode_value))
print("Variance:", variance_value)
print("Standard Deviation:", std_dev_value)
print("Skewness:", skewness_value)
print("Kurtosis:", kurtosis_value)
```
